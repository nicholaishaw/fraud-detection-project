```python
#Import dependencies
import numpy as np
import pandas as pd
from pathlib import Path
from datetime import datetime
```

# Data Export and Cleaning

```python
#Read in the csv file to Pandas dataframe
path = Path('resources\\raw_data\\fraud_data.csv')
fraud_df = pd.read_csv(path)
fraud_df
```

```python
#Get a list of the columns
fraud_df.columns
```

```python
#Rename 'Unamed: 0' to 'transaction id'
fraud_df.rename(columns={'Unnamed: 0' : 'transaction id'}, inplace=True)
fraud_df
```

```python
#Rename amt to amount
fraud_df.rename(columns={'amt' : 'amount'}, inplace=True)
fraud_df
```

```python
#Convert the date and time in the column 'trans_date_trans_time'
fraud_df['trans_date_trans_time'] = pd.to_datetime(fraud_df['trans_date_trans_time'])

#Split the 'trans_date_trans_time' column into two columns: date and time
fraud_df['date'] = fraud_df['trans_date_trans_time'].dt.date
fraud_df['time'] = fraud_df['trans_date_trans_time'].dt.time
fraud_df
```

```python
#Drop the 'trans_date_trans_time' since it's been split into separate date and time columns
fraud_df.drop(columns='trans_date_trans_time', inplace=True)
```

```python
#In order to convert the date of birth into age, we need the following code:

#Convert the date of birth column ('dob') to a datetime object 
fraud_df['dob'] = pd.to_datetime(fraud_df['dob'])

#Get the current date and time with the datetime.now() function
current_date = datetime.now()

#Make a new column in our dataframe named 'age' that will subtract our current date and the dob
#"astype('<m8[Y]')" simply means to convert the date change (current date - date of birth, in this case) to a NumPy object
fraud_df['age'] = (current_date - fraud_df['dob']).astype('<m8[Y]')

#Drop the 'dob' column
fraud_df.drop(columns='dob', inplace=True)


fraud_df
```

```python
#Get the columns for reordering
fraud_df.columns
```

```python
#Reorder the columns
fraud_df = fraud_df[['transaction id', 'date', 'time', 'cc_num', 'merchant', 'category', 'amount', 'first',
       'last', 'age', 'gender', 'street', 'city', 'state', 'zip', 'lat', 'long',
       'city_pop', 'job', 'trans_num', 'unix_time', 'merch_lat',
       'merch_long', 'is_fraud']]
```

```python
#Drop the columns that are not features of the model
fraud_df.drop(columns=['trans_num', 'cc_num', 'first', 'last', 'zip', 'merch_lat', 'merch_long', 'street', 
                       'unix_time', 'merchant', 'job', 'city', 'date', 'lat', 'long'], inplace=True)
fraud_df
```

```python
#Export the clean data to a separate folder
fraud_df.to_csv('resources\\clean_data\\fraud_clean_data.csv', index=False)
```

# SQLAlchemy Export

Below is the export of the 'fraud_df' in Pandas to the 'FraudDetection' SQL database for this project. This will automatically create a new table 'fraud_data' if it does not already exist. 

When creating the engine, Replace 'PASSWORD' with your password used to access the project database in pgAdmin4, and replace "IP_ADDRESS" with the IP address or hostname of the server where the PostgreSQL database is running.

To obtain your IP address, type either "ipconfig" for windows or "ifconfig" for mac in the terminal and obtain the series of numbers that corresponds to the "IPv4 Address" in the output.

```python
#Import sqlalchemy
from sqlalchemy import create_engine

# Create the engine to interact with the SQL database. Replace 'PASSWORD' with the password your password.\
# and "IP_ADDRESS" with your IP address
engine = create_engine('postgresql://postgres:PASSWORD@IP_ADDRESS/FraudDetection')

# Load the dataframe into the SQL Database
fraud_df.to_sql('fraud_data', engine, index=False, if_exists='replace')
```
